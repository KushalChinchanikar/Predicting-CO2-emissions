```python
df = pd.read_csv("Fuel Consumption Ratings 2021.csv")
```

```python
df
```

```python
df = df.iloc[: , :15] #selecting first 15 columns as others are empty
df = df.iloc[:917]
df
```

```python
sel_df = df[['Engine_Size','Cylinders','Fuel_Consumption_Comb','CO2_Emissions']]
sel_df
```

```python
plt.scatter(sel_df.Fuel_Consumption_Comb, sel_df.CO2_Emissions, color="red")
plt.xlabel("Fuel Consumption Comb")
plt.ylabel("CO2 Emissions")
plt.show()
```

```python
msk = np.random.rand(len(sel_df)) < 0.8   # 80% of the rows selected randomly for mask 
train = sel_df[msk] # 80% rows
test = sel_df[~msk] # 20% rows
```

```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()
train[['Engine_Size']]

x_train = np.asanyarray(train[['Engine_Size']])
y_train = np.asanyarray(train[['CO2_Emissions']])

lreg.fit(x_train, y_train)

print('coefficient is ', lreg.coef_)
print('intercept is ',lreg.intercept_)
```

```python
plt.scatter(train.Engine_Size, train.CO2_Emissions, color="Green")
plt.plot(x_train, lreg.coef_[0][0] * x_train + lreg.intercept_[0],'-r')
plt.xlabel('Engine Size')
plt.ylabel("Co2 Emissions")
```

```python
sel_df1 = df[['Engine_Size','Cylinders','Fuel_Consumption_Comb', 'CO2_Emissions']]
sel_df1.head()
```

```python
from sklearn.metrics import r2_score

x_test = np.asanyarray(test[['Engine_Size']])
y_test = np.asanyarray(test[['CO2_Emissions']])
y_test_predicted = lreg.predict(x_test)

print("Mean absolute error - %.2f" % np.mean(np.absolute(y_test_predicted - y_test)))
print("Mean Square Error - %.2f" % np.mean((y_test_predicted - y_test)**2))
print("R2 score is %.2f" % r2_score(y_test, y_test_predicted)) 
```
